```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(0)
x = np.random.randint(18,70,size=100)

plt.hist(x, density=True, bins=20,edgecolor="black") # density=False would make counts
plt.ylabel('Age')
plt.xlabel('Frequency')
plt.title('Distribution of Ages in a Population')
plt.show()
```
